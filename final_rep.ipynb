```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
#from datetime import datetime
```

#### Creating a ticker object for Tesla Stocks and obtaining the informations from the ticker

```python
tsla = yf.Ticker("TSLA")
tsla.info
```

#### Creating the dataframe for TSLA prices from its beginning

```python
hist_df = tsla.history(period="max")
type(hist_df)
```

#### Modifying the dataframe with a beginning date 2020

```python
hist_df["2020":]
```

#### Creating a csv file of the dataframe

```python
hist_df.to_csv("hist_df.csv")
```

#### Creating a dataframe with a start date 2020

```python
hist_df_2020 = hist_df["2020":]
hist_df_2020
```

#### All time TSLA Stocks dataframe

```python
hist_df
```

#### In order to plot the data that we obtained from the dataframes

```python
%matplotlib inline
```

#### Subplotting the data of its open, close, high and low prices of all times

```python
ax1 = plt.subplot(511)
hist_df["Open"].plot(figsize=(20,30), label="TSLA Open", title="TSLA Open")
plt.legend(loc="upper center")
ax2 = plt.subplot(512)
hist_df["Close"].plot(figsize=(20,30), label="TSLA Close", title="TSLA Close")
plt.legend(loc="upper center")
ax3 = plt.subplot(513)
hist_df["High"].plot(figsize=(20,30), label="TSLA High", title="TSLA High")
plt.legend(loc="upper center")
ax4 = plt.subplot(514)
hist_df["Low"].plot(figsize=(20,30), label="TSLA Low", title="TSLA Low")
plt.legend(loc="upper center")
ax5 = plt.subplot(515)
hist_df["Volume"].plot(figsize=(20,30), label="TSLA Volume", title="TSLA Volume")
plt.legend(loc="upper center")
plt.tight_layout()
plt.show()
```

#### Subplotting the data since 2020 of its open, close, high and low prices

```python
ax1 = plt.subplot(511)
hist_df_2020["Open"].plot(figsize=(20,30), label="TSLA Open")
ax2 = plt.subplot(512)
hist_df_2020["Close"].plot(figsize=(20,30), label="TSLA Close")
ax3 = plt.subplot(513)
hist_df_2020["High"].plot(figsize=(20,30), label="TSLA High")
ax4 = plt.subplot(514)
hist_df_2020["Low"].plot(figsize=(20,30), label="TSLA Low")
ax5 = plt.subplot(515)
hist_df_2020["Volume"].plot(figsize=(20,30), label="TSLA Volume")
plt.show()
```

#### Creating a candlestick plot of its open, high, close and low prices from 2020 to date

```python
from plotly.offline import plot
fig = go.Figure(data=[go.Candlestick(x=hist_df_2020.index,
                open=hist_df_2020['Open'],
                high=hist_df_2020['High'],
                low=hist_df_2020['Low'],
                close=hist_df_2020['Close'])])

plot(fig)
```

#### Plotting moving average of its close prices for 30 days and its close prices

```python
"""rolling methodu ile son 30 günün avg'ını movelayarak hesaplayabiliyoruz"""
plt.figure(figsize=(20,12))
hist_df_2020['Close'].loc['2020-01-01':].rolling(window=30).mean().plot(label='MA 30')
hist_df_2020['Close'].loc['2020-01-01':].plot(label='TSLA Close')
plt.legend()
plt.show()
```

```python
hist_df_2020['Close'].loc['2020-01-01':].rolling(window=30).mean()
```

#### Plotting the data of from the last month of 2021 of its open, close, high and low prices by indicating their labels and colors

```python
plt.figure(figsize=(25,12))
plt.plot(hist_df_2020["2021-12":]["Open"], "b--", label="Open")
plt.plot(hist_df_2020["2021-12":]["Close"], c="y", label="Close")
plt.plot(hist_df_2020["2021-12":]["High"], c="m", label="High")
plt.plot(hist_df_2020["2021-12":]["Low"], c="r", label="Low")
plt.legend()
plt.show()
```

#### Creating a stem plot of delta values of open and close prices of 2020 dataframe

```python
plt.figure(figsize=(25,12))
plt.stem(hist_df_2020.index, (hist_df_2020["Open"]-hist_df_2020["Close"]))
plt.title("Difference between Open and Close daily")
plt.show()
```
